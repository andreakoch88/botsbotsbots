```python
tweets_list = []
for tweet in query:
    tweets_list.append((tweet.fullname,
                        tweet.id,
                        tweet.likes,
                        tweet.replies,
                        tweet.retweets,
                        tweet.text,
                        tweet.timestamp,
                        tweet.user))
```
